### Importación de datos



```python
import pandas as pd

url_01 = "https://raw.githubusercontent.com/AlanSebastianArce/Challenge-Alura-Store-G9-ONE/refs/heads/main/alura-store-database/tienda_01.csv"
url_02 = "https://raw.githubusercontent.com/AlanSebastianArce/Challenge-Alura-Store-G9-ONE/refs/heads/main/alura-store-database/tienda_02.csv"
url_03 = "https://raw.githubusercontent.com/AlanSebastianArce/Challenge-Alura-Store-G9-ONE/refs/heads/main/alura-store-database/tienda_03.csv"
url_04 = "https://raw.githubusercontent.com/AlanSebastianArce/Challenge-Alura-Store-G9-ONE/refs/heads/main/alura-store-database/tienda_04.csv"

tienda_01 = pd.read_csv(url_01)
tienda_02 = pd.read_csv(url_02)
tienda_03 = pd.read_csv(url_03)
tienda_04 = pd.read_csv(url_04)

tienda_01.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de cada tienda
ingreso_total_01 = tienda_01['Precio'].sum()
ingreso_total_02 = tienda_02['Precio'].sum()
ingreso_total_03 = tienda_03['Precio'].sum()
ingreso_total_04 = tienda_04['Precio'].sum()
```

```python
#Mostrar ingresos de las tiendas
print('Ingresos Totales por cada Tienda:')
print('\n******************************')
print(f'Tienda 1 = ${ingreso_total_01}')
print('\n******************************')
print(f'Tienda 2 = ${ingreso_total_02}')
print('\n******************************')
print(f'Tienda 3 = ${ingreso_total_03}')
print('\n******************************')
print(f'Tienda 4 = ${ingreso_total_04}')
print('\n******************************')
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda
print('Cantidad de productos vendidos en cada una de las categorias de cada una de las tiendas\n')
print('******************************')
print('"Tienda 1"')
print('******************************')
print(tienda_01['Categoría del Producto'].value_counts())
print('******************************')
print('"Tienda 2"')
print('******************************')
print(tienda_02['Categoría del Producto'].value_counts())
print('******************************')
print('"Tienda 3"')
print('******************************')
print(tienda_03['Categoría del Producto'].value_counts())
print('******************************')
print('"Tienda 4"')
print('******************************')
print(tienda_04['Categoría del Producto'].value_counts())
print('******************************')
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de clientes por tienda
print('Calificación promedio de clientes por tienda\n')
print('*************************************')
print(f'Calificación promedio Tienda 1 = {tienda_01['Calificación'].mean():.2f}')
print(f'Calificación promedio Tienda 2 = {tienda_02['Calificación'].mean():.2f}')
print(f'Calificación promedio Tienda 3 = {tienda_03['Calificación'].mean():.2f}')
print(f'Calificación promedio Tienda 4 = {tienda_04['Calificación'].mean():.2f}')
print('\n*************************************')
```

# 4. Productos más y menos vendidos

```python
print('****************************************************')
print('Productos más y menos vendidos por tienda\n')
print('****************************************************')
for nombre, df in [("Tienda 1", tienda_01), ("Tienda 2", tienda_02),
                   ("Tienda 3", tienda_03), ("Tienda 4", tienda_04)]:
    ventas = df['Producto'].value_counts()
    print(f"\n{nombre}:")
    print(f"  Más vendido:   {ventas.index[0]} ({ventas.iloc[0]} ventas)")
    print(f"  Menos vendido: {ventas.index[-1]} ({ventas.iloc[-1]} ventas)")
print('\n**************************************************')
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
print('*************************************************************************************')
print('A continuación se detalla el "Costo de envío promedio" para cada una de las tiendas.')
print('*************************************************************************************')
print(f'"Costo de envío promedio" para la Tienda 1: ${tienda_01['Costo de envío'].mean():.2f}')
print(f'"Costo de envío promedio" para la Tienda 2: ${tienda_02['Costo de envío'].mean():.2f}')
print(f'"Costo de envío promedio" para la Tienda 3: ${tienda_03['Costo de envío'].mean():.2f}')
print(f'"Costo de envío promedio" para la Tienda 4: ${tienda_04['Costo de envío'].mean():.2f}')
print('\n*************************************************************************************')
```

# 6. Visualización

```python
import matplotlib.pyplot as plt
import numpy as np

tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
tiendas_dfs = [tienda_01, tienda_02, tienda_03, tienda_04]
colores = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
```

```python
print('\n\nGRÁFICO 1 - BARRAS: Ingresos totales por tienda')
# ─────────────────────────────────────────────
# GRÁFICO 1 - BARRAS: Ingresos totales por tienda
# ─────────────────────────────────────────────
ingresos = [df['Precio'].sum() / 1_000_000 for df in tiendas_dfs]

fig, ax = plt.subplots(figsize=(8, 5))
barras = ax.bar(tiendas_nombres, ingresos, color=colores, edgecolor='white', linewidth=1.2)

for barra, valor in zip(barras, ingresos):
    ax.text(barra.get_x() + barra.get_width() / 2, barra.get_height() + 0.5,
            f'${valor:.1f}M', ha='center', va='bottom', fontweight='bold', fontsize=10)

ax.set_title('Ingresos Totales por Tienda', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Ingresos (millones $)', fontsize=11)
ax.set_ylim(0, max(ingresos) * 1.15)
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
print('\n\nGRÁFICO 2 - BARRAS AGRUPADAS: Ventas por categoría')
# ─────────────────────────────────────────────
# GRÁFICO 2 - BARRAS AGRUPADAS: Ventas por categoría
# ─────────────────────────────────────────────
categorias = tienda_01['Categoría del Producto'].unique()
x = np.arange(len(categorias))
ancho = 0.2

fig, ax = plt.subplots(figsize=(12, 6))
for i, (df, nombre, color) in enumerate(zip(tiendas_dfs, tiendas_nombres, colores)):
    conteos = df['Categoría del Producto'].value_counts().reindex(categorias, fill_value=0)
    ax.bar(x + i * ancho, conteos.values, width=ancho, label=nombre, color=color, edgecolor='white')

ax.set_title('Ventas por Categoría de Producto', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Cantidad de ventas', fontsize=11)
ax.set_xticks(x + ancho * 1.5)
ax.set_xticklabels(categorias, rotation=20, ha='right', fontsize=9)
ax.legend(title='Tienda')
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
print('\n\nGRÁFICO 3 - LÍNEAS: Calificación promedio y costo de envío')
# ─────────────────────────────────────────────
# GRÁFICO 3 - LÍNEAS: Calificación promedio y costo de envío
# ─────────────────────────────────────────────
calificaciones = [df['Calificación'].mean() for df in tiendas_dfs]
envios = [df['Costo de envío'].mean() / 1000 for df in tiendas_dfs]

fig, ax1 = plt.subplots(figsize=(8, 5))

# Eje izquierdo - Calificación
ax1.plot(tiendas_nombres, calificaciones, color='#4C72B0', marker='o',
         linewidth=2.5, markersize=8, label='Calificación promedio')
ax1.set_ylabel('Calificación promedio (1-5)', fontsize=11, color='#4C72B0')
ax1.set_ylim(3.5, 4.5)
ax1.tick_params(axis='y', labelcolor='#4C72B0')

# Eje derecho - Envío
ax2 = ax1.twinx()
ax2.plot(tiendas_nombres, envios, color='#C44E52', marker='s',
         linewidth=2.5, markersize=8, linestyle='--', label='Costo envío promedio')
ax2.set_ylabel('Costo de envío promedio (miles $)', fontsize=11, color='#C44E52')
ax2.tick_params(axis='y', labelcolor='#C44E52')

ax1.set_title('Calificación Promedio vs Costo de Envío por Tienda', fontsize=13, fontweight='bold', pad=15)
ax1.grid(axis='y', linestyle='--', alpha=0.4)
ax1.spines[['top']].set_visible(False)

# Leyenda combinada
lineas1, labels1 = ax1.get_legend_handles_labels()
lineas2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lineas1 + lineas2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.show()
```

## 3. Conclusión y Recomendación

### 🏪 Tienda recomendada para vender: Tienda 1

Tras evaluar los cinco factores analizados, se recomienda que el Sr. Juan
**venda la Tienda 1**, por las siguientes razones:

- 💰 **Mayor ingreso total**: La Tienda 1 genera la mayor facturación
  ($1,150,880,400), lo que la convierte en el activo más valioso del portafolio.
  Venderla representaría el mayor retorno económico posible.

- 📦 **Costo de envío más elevado**: Con un promedio de $26,018 por envío,
  es la tienda con mayor costo logístico, lo cual podría erosionar la rentabilidad
  a futuro si no se optimiza.

- ⭐ **Calificación más baja**: Con 3.98 puntos, presenta la menor satisfacción
  del cliente entre las cuatro tiendas, una señal de alerta sobre su experiencia
  de compra que podría volverse un problema estructural.

### ✅ Razonamiento estratégico

Vender la Tienda 1 en este momento es estratégicamente inteligente: **se vende
en su punto más alto de ingresos**, antes de que los problemas de satisfacción
y costos logísticos impacten negativamente su valor. Las Tiendas 2, 3 y 4
muestran mejor equilibrio entre calificación, costo de envío y volumen de ventas,
por lo que conservarlas representa una base más sólida y sostenible para el negocio.
